```python
%matplotlib inline
from skimage import io
import numpy as np
from matplotlib import pyplot as plt

# Computes the gradient for a 1D list
def compute1DGradient( v ):
    
    v_len = v.shape[0]
    gradients = np.empty(len(v))
    
    # Calculate gradient for the outer elements
    gradients[0] = (v[1] - v[0]) / 1.0
    gradients[v_len-1] = (v[v_len-1] - v[v_len-2]) / 1.0

    # Calculate gradients for the inner elements
    for i in range(1, v_len-1):
        
        # Do Floating Point division! 
        # The distance between two elements is now 2.0
        gradient = (v[i+1] - v[i-1]) / 2.0
        gradients[i] = gradient
    
    return gradients


# Computes the gradient for a 2D list
def compute2DGradient( m ):
  
    gradients_x = np.empty([m.shape[0],m.shape[1]])
    gradients_y = np.empty([m.shape[1],m.shape[0]])
    
    for row in range(0, m.shape[0]):
        row_m = m[row]
        gradient_row_m = compute1DGradient(row_m)
        gradients_x[row] = gradient_row_m
    
    for col in range(0, m.shape[1]):
        col_m = m[:,col]
        gradient_col_m = compute1DGradient(col_m)
        gradients_y[col] = gradient_col_m
    
    # We can do the transpose set to have the same format as np.gradient() returns
    return np.array([np.transpose(gradients_y),gradients_x])



# TODO: Verify if this correct
# mag = sqrt( gx^2 + gy^2 )    
def magnitude(m):
    gy2 = np.power(m[0], 2)
    gx2 = np.power(m[1], 2)
    mag = np.sqrt(np.add(gx2,gy2))
    
    return mag


# TODO Verify if this is correct
# theta = tan-1( gy / gx )  
def orientation(m):
    gy = m[0]
    gx = m[1]
    theta = np.rad2deg(np.arctan2(gy, gx))
    
    return theta


def applyThreshold(m, min, max):
    
    new_m = np.empty(m.shape)

    for i in range(0,m.shape[0]):
        for j in range(0,m.shape[1]):
            if (m[i][j] < min) or (m[i][j] > max):
                new_m[i][j] = 1
            else:
                new_m[i][j] = 0
                
    return new_m

# test_matrix = np.array([[1, 2, 4, 7, 11, 16],[4, 9, 4, 11, 1, 4],[4, 5, 2, 1, 9, 2]])
# print( np.gradient(test_matrix) )


# We can also test it with a real image

img = io.imread("testimage.png", as_grey=True)

gradients = compute2DGradient(img)
magnitudes = magnitude(gradients)

plt.imshow(applyThreshold(gradients[0],-0.1,0.1), cmap='gray') # Y-Direction
plt.figure()
plt.imshow(applyThreshold(gradients[1],-0.1,0.1), cmap='gray') # X-Direction
plt.figure()
plt.imshow(magnitudes, cmap='gray') # Magnitudes
plt.figure()
plt.imshow(applyThreshold(magnitudes,-0.1,0.1), cmap='gray') # Magnitudes with Threshold
```
